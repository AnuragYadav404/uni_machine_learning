### 1.2.1 Inertia

Inertia is the sum of squared distances of samples to their closest cluster center. Essentially, it measures how internally coherent the clusters are.

In K-means clustering, the inertia is calculated as the sum of squared distances between each data point and the centroid of the cluster it belongs to. 
$$ \text{Inertia} = \sum_{i=1}^{n} \min_{\mu_j \in C} \left( || x_i - \mu_j ||^2 \right)
$$

Where:

$ n $ is the number of samples (data points) in your dataset
$\mu_j $ represents the centroid of cluster jj from the set of centroids $C$
$x_i$ is the $i$-th data point
$|| x_i - \mu_j ||^2$ is the squared Euclidean distance between the data point $x_i$  and the  cluster centroid $\mu_j$ 

### Task 7: Compute the Inertia Score for the Two Previous K-means Clustering Tasks, in Dataset 1 and  2

## 1.2.2 Silhouette Coefficient 

The silhouette coefficient gives an indication of how close each point in one cluster is to points in the neighboring clusters and thus is a measure of how appropriately the data has been clustered

The silhouette coefficient for a single data point is calculated as follows:

$$
s(i) = \frac{b(i) - a(i)}{\max\{a(i), b(i)\}}
$$

Where for each data point $i$:

$a(i)$ is the average distance from the $i$-th data point to the other points in the same cluster.

$b(i)$ is the smallest average distance from the $i$-th data point to points in a different cluster, minimized over clusters.

The silhouette score for a set of data points is the average of the silhouette coefficient for each point.


### Task 8: Compute the Silhuette Score for the Two Previous K-means Clustering Tasks

```python
# to do : compute and print the metrics for the 2 datasets
from sklearn.metrics import silhouette_score
silhouette_avg_1 = silhouette_score(X1, kmeans1.predict(X1))
print("Silhouette dist for data 1: ", silhouette_avg_1)
silhouette_avg_2 = silhouette_score(X2, kmeans2.predict(X2))
print("Silhouette dist for data 2: ", silhouette_avg_2)
```

## 1.2.3 CalinskiHarabasz (CH) Score

Given a data set of n points: {x1, ..., xn}, and the assignment of these points to k clusters {C1, ..., Ck}, the CalinskiHarabasz (CH) score is defined as the ratio of the between-cluster separation (BCSS) to the within-cluster dispersion (WCSS), normalized by their number of degrees of freedom

$$ CH = \frac{\frac{BCSS}{(k - 1)}}{\frac{WCSS}{(n - k)}} $$

BCSS (Between-Cluster Sum of Squares) is the weighted sum of squared Euclidean distances between each cluster centroid (mean) and the overall data centroid (mean)

$$ BCSS = \sum_{i=1}^{k} n_i \left\| c_i - c \right\|^2 $$


where ni is the number of points in cluster Ci, ci is the centroid of Ci, and c is the overall centroid of the data

WCSS (Within-Cluster Sum of Squares) is the sum of squared Euclidean distances between the data points and their respective cluster centroids

$$ WCSS = \sum_{i=1}^{k} \sum_{x \in C_i} \left\| x - c_i \right\|^2 $$


### Task 9: Compute the Calinski Harabasz Score for the Two Previous K-means Clustering Tasks

## 1.2.4 Davies-Bouldin Score

The score is defined as the average similarity measure of each cluster with its most similar cluster, where similarity is the ratio of within-cluster distances to between-cluster distances

How is calculated?

- For each cluster $i$, compute the average distance between each point in the cluster and the centroid of that cluster. This average distance is a measure of the cluster's scatter and is denoted by $S_i$.

- For each pair of clusters $i$ and $j$, calculate the distance $d(i,j)$ between the centroids of the two clusters

- The similarity $R_{ij}$ between two clusters $i$ and $j$ is defined as the ratio of their scatters and the centroid distance:

$$ R_{ij} = \frac{S_i + S_j}{d(i, j)}$$

The Davies-Bouldin index for the entire clustering is then calculated by taking the average of the maximum similarity of each cluster with its most similar cluster:

$$ DB = \frac{1}{k} \sum_{i=1}^{k} \max_{j \neq i} R_{ij}$$

Where:

$k$ is the total number of clusters
    

### Task 10: Compute the Davies Bouldin Score for the Two Previous K-means Clustering Tasks

## 2. Image segmentation with K-means

An image segmentation task involves dividing an image into distinct regions based on the similarities of pixel values, usually their color. This can be used for various applications like object detection, medical imaging, or background separation. 

*Clustering* can be used for this task too. In this programming task, K-means is used for segmenting an image. The process can be described in the following steps:

- Download the image data to be segmented to the working directory
- Reshape the image data: The image is reshaped into a 2D array where each row represents a pixel and the columns represent the color channels (e.g., RGB). This step is necessary because k-means clustering in this context is applied on a per-pixel basis
- Apply K-means Clustering
- Post-process: After clustering, each pixel's color is replaced with the centroid color of the cluster it belongs to. This results in segmenting the image into regions of similar colors
- Reconstruct the mage: The 2D array is reshaped back into the original image dimensions, but now with the segmented colors

    Result: The final output is a segmented image where each region represents a cluster of pixels with similar colors or intensities.
